```python
import pymysql.cursors
connection = pymysql.connect(host="localhost", 
                     user="root",   
                     passwd="root", 
                     db="sakuratrip",   
                     charset='utf8mb4',
                     cursorclass=pymysql.cursors.DictCursor)
```

```python
with connection.cursor() as cursor:
        sql = "SELECT * FROM sites"
        cursor.execute(sql)
        result = cursor.fetchall()
        dataset = []
        temp = {}
        i = 0
        for row in result:
            temp[i] = row
            dataset.append(temp[i])
            i = i + 1
        print('\n')
        print(dataset)
        print('\n')
```
